# Voronoi Ponderado aplicado a soccer
Este notebook es un tutorial para aprender a utilizar la visualización de Voronoi ponderado, se basa en el código 
`voronoi_weighted.py`.
* Q&A erika.munoz.datascientist@gmail.com

Importamos la bibloteca y algunas de sus funciones

```python
from voronoi_weighted import weighted_voronoi,plot_voronoi, plot_pitch
import matplotlib.pyplot as plt
%matplotlib inline
```

Podemos dibujar el campo de soccer

```python
plot_pitch(lc = 'black', lw = 2, xPitchSize = 100, yPitchSize = 60, scale = 1, axis = 'on')
```

Podemos variar el tamaño del campo, las medidas se encuentran dadas en metros, así como el color de linea 
`lc = green `, el grosor de las líneas  
`lw = 5` y tambien ocultar los ejes numerados con `axis = off`.

```python
plot_pitch(lc = 'green', lw = 4, xPitchSize = 120, yPitchSize = 68, scale = 1, axis = 'off')
```

Damos las posiciones de los jugadores de ambos equipos como una lista de python

```python
jugadores_local = [(10,30),(35,21),(34,30),(36,37),(38,39),(40,40),(40,32),(55,35),(56,50),(55,58),(60,50)]
jugadores_visita = [(34,20),(37,34),(37,40),(40,50),(55,15),(54,32),(56,45),(58,56),(70,28),(72,45),(80,35)] 
```

Podemos dibujar a los jugadores en el campo

```python
plot_pitch(lc = 'black', lw = 2, xPitchSize = 100, yPitchSize = 60, scale = 1, axis = 'on')
for x,y in jugadores_local:
    plt.scatter(x, y, marker = 'o', color = 'red')
for x,y in jugadores_visita:
    plt.scatter(x, y, marker = 'o', color = 'blue')
```

Ahora calculamos la matriz de Voronoi ponderado y la dibujamos

```python
wv = weighted_voronoi(jugadores_local, jugadores_visita, resolution = 200, beta = 2)
plot_voronoi(wv, figsize = (20,12), lc = 'black', interpolation = 'none', cmap = 'seismic', scale = 1, axis = 'off')
```
